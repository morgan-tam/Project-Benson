```python
import pandas as pd
from datetime import datetime as dt
%matplotlib inline
```

**Hypothesis:** 
Rainy weather may or may not have a significant impact on ridership on busiest stations

**Considerations:**
Ridership impacts may vary between stations, so we run comparisons only between stations of interest.Date and time are still most important predictors of ridership. Therefore we will need to hold these values constant in our comparisons.

**Variables**
For this set, we only care about unique turnstiles, unique stations, date, weekend/weekday, transits, and weather


```python
df = pd.read_csv('/Users/Collin/DS/Project-Benson/data_daily_weather.csv')
df['DATE'] = pd.to_datetime(df['DATE'])
df['DAY_OF_WEEK'] = df['DATE'].dt.weekday_name
```

**Step 1: Identify Priority Staions**

```python
"""First, we identify the busiest stations by cumulative ridership"""
transit_station = df.groupby('UN_STATION', as_index=False).sum()
busiest_stations = transit_station.sort_values('TRANSITING', ascending=False).head(20)
```

```python
"""Now we select data from the original table where stations matches one of the above 20"""
df_b = pd.DataFrame()

for station in busiest_stations.UN_STATION:
    df_b = df_b.append(df.loc[df['UN_STATION'] == station])

df_b
```

**Step 2: Create Dataframes for Rainy and Sunny Days**

```python
"""Here, we decide that a rainy day is a day that receives a half inch of rain"""
rainy_df = df_b.loc[df_b['PRCP'] >=.5 ]
sunny_df = df_b.loc[df_b['PRCP'] <.5 ]
```

```python
"We see that we have rainy days each day of the week"
print ("# of RAINY DAYS: %r" %(len(rainy_df.DAY_OF_WEEK.unique())))
print ("-------------------------")
print ("VALUE COUNTS")
rainy_df.DAY_OF_WEEK.value_counts()
```

**Step 3: Group by Weekdays and Weekends**

```python
Weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
Weekends = ["Saturday", "Sunday"]
```

```python
'''Here, we assign a Weekday or Weekend variable to each turnstile-date entry in rainy and sunny dataframes'''

def is_weekday(x):
    if x in Weekdays:
        return 1
    else:
        return 0

rainy_df["WEEKDAY"] = rainy_df["DAY_OF_WEEK"].map(is_weekday)
sunny_df["WEEKDAY"] = sunny_df["DAY_OF_WEEK"].map(is_weekday)
```

```python
"""ONE CAVEAT: we have many many more sunny turnstile data than rainy turnstile data"""
print("RAINY DAYS by WEEKDAY")
print(rainy_df.WEEKDAY.value_counts())

print ('-----------------')
print("SUNNY DAYS by WEEKDAY")
print (sunny_df.WEEKDAY.value_counts())
```

```python
"""lets split weekday and weekend data into seperate frames"""

rainy_weekday = rainy_df.loc[rainy_df['WEEKDAY']==1]
rainy_weekend = rainy_df.loc[rainy_df['WEEKDAY']==0]

sunny_weekday = sunny_df.loc[sunny_df['WEEKDAY']==1]
sunny_weekend = sunny_df.loc[sunny_df['WEEKDAY']==0]
```

```python
rainy_weekday.head()
```

```python
"""SInce our dataframes are still seperated by turnstile, we need to group these together into stations
However, we have to take the mean of each turnstile before summing because we have many more turnstile entries 
in the sunny datasset than the rainy dataset"""

rainy_weekday_turnstile_avgs = rainy_weekday.groupby(["UN_STATION", "C/A", "UNIT" ,"SCP"], as_index = False).mean()
rainy_weekday_station_avgs = rainy_weekday_turnstile_avgs.groupby(["UN_STATION"], as_index = False).sum()[["UN_STATION", "TRANSITING"]]

rainy_weekend_turnstile_avgs = rainy_weekend.groupby(["UN_STATION", "C/A", "UNIT" ,"SCP"], as_index = False).mean()
rainy_weekend_station_avgs = rainy_weekend_turnstile_avgs.groupby(["UN_STATION"], as_index = False).sum()[["UN_STATION", "TRANSITING"]]

sunny_weekday_turnstile_avgs = sunny_weekday.groupby(["UN_STATION", "C/A", "UNIT" ,"SCP"], as_index = False).mean()
sunny_weekday_station_avgs = sunny_weekday_turnstile_avgs.groupby(["UN_STATION"], as_index = False).sum()[["UN_STATION", "TRANSITING"]]

sunny_weekend_turnstile_avgs = sunny_weekend.groupby(["UN_STATION", "C/A", "UNIT" ,"SCP"], as_index = False).mean()
sunny_weekend_station_avgs = sunny_weekend_turnstile_avgs.groupby(["UN_STATION"], as_index = False).sum()[["UN_STATION", "TRANSITING"]]
```

```python
"""Here, we just add rainy and sunny transits to a single dataframe"""

rainy_weekday_station_avgs["RAINY_TRANSITS"] = rainy_weekday_station_avgs["TRANSITING"]
rainy_weekday_station_avgs["SUNNY_TRANSITS"] = sunny_weekday_station_avgs["TRANSITING"]

weekday = rainy_weekday_station_avgs[["UN_STATION", "RAINY_TRANSITS" ,"SUNNY_TRANSITS"]]

rainy_weekend_station_avgs["RAINY_TRANSITS"] = rainy_weekend_station_avgs["TRANSITING"]
rainy_weekend_station_avgs["SUNNY_TRANSITS"] = sunny_weekend_station_avgs["TRANSITING"]

weekend = rainy_weekend_station_avgs[["UN_STATION", "RAINY_TRANSITS" ,"SUNNY_TRANSITS"]]
```

```python
"""Here we set up a difference of the mean calc which we can use in AB testing in the future"""
weekday['RAINY - SUNNY'] = weekday["RAINY_TRANSITS"] - weekday["SUNNY_TRANSITS"]
weekend['RAINY - SUNNY'] = weekend["RAINY_TRANSITS"] - weekend["SUNNY_TRANSITS"]

"""We still need to do the standard deviation of each station for all dates"""
```
